# Iraq World Development Indicators - Data Import

This notebook imports the World Bank World Development Indicators dataset for Iraq.

```python
import pandas as pd
import os

# Define the data directory
data_dir = 'API_IRQ_DS2_en_csv_v2_16712'

# File paths
data_file = os.path.join(data_dir, 'API_IRQ_DS2_en_csv_v2_16712.csv')
country_meta_file = os.path.join(data_dir, 'Metadata_Country_API_IRQ_DS2_en_csv_v2_16712.csv')
indicator_meta_file = os.path.join(data_dir, 'Metadata_Indicator_API_IRQ_DS2_en_csv_v2_16712.csv')
```

## 1. Import Main Data File

The main CSV contains World Development Indicators for Iraq across multiple years. It has metadata rows at the top (Data Source, Last Updated Date) which we skip.

```python
# Read the main data file, skipping the first 4 metadata rows
df = pd.read_csv(data_file, skiprows=4)

# Display basic info
print('Shape:', df.shape)
print('\nColumns:', df.columns.tolist())
print('\nFirst few rows:')
df.head()
```

## 2. Import Country Metadata

```python
# Read country metadata
df_country = pd.read_csv(country_meta_file)

print('Shape:', df_country.shape)
print('\nColumns:', df_country.columns.tolist())
df_country.head()
```

## 3. Import Indicator Metadata

```python
# Read indicator metadata
df_indicator = pd.read_csv(indicator_meta_file)

print('Shape:', df_indicator.shape)
print('\nColumns:', df_indicator.columns.tolist())
df_indicator.head()
```

## 4. Quick Data Preview

Check how many indicators are available and preview some data types.

```python
# List all indicators in the dataset
print(f"Total indicators: {df['Indicator Name'].nunique()}")
print('\nIndicator Names:')
for ind in df['Indicator Name'].unique():
    print(f'  - {ind}')
```

```python
# Check data types and non-null counts for the main dataframe
df.info()
```

```python
# Quick preview of year columns (1960-2025)
year_columns = [col for col in df.columns if col.isdigit()]
print(f'Number of year columns: {len(year_columns)}')
print(f'Year range: {year_columns[0]} - {year_columns[-1]}')
```
